```python
from sklearn.preprocessing import LabelEncoder

def convertObjToInt(df, cols):
    encoders = {}
    for col in cols:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        encoders[col] = le  # save the encoder for future use
    return df, encoders

# Columns to convert
columns = ['Make', 'FuelType', 'Transmission', 'Drivetrain']

# Apply conversion
df, encoders = convertObjToInt(df, columns)
```

```python
import pandas as pd
df = pd.read_csv('processed_data.csv')
# df.info()
df.columns
```

```python
X = df.drop('Price', axis=1)   # Dependant columns / Input columns
y = df['Price']                # Resultant columns / Output column
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.2, shuffle=False) # 80:20, 70:30, 60:40
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd
import numpy as np


cat_model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.05,
    depth=6,
    verbose=0,
    random_state=42
)

```

```python
cat_model.fit(train_X, train_y)
```

```python
y_pred = cat_model.predict(test_X)
y_pred
```

```python
r2 = r2_score(test_y, y_pred)
accuracy = round(r2 * 100, 2) 
rmse = np.sqrt(mean_squared_error(test_y, y_pred))

print(f"R2 Score: {accuracy} %")
print(f"Prediction Accuracy: {accuracy} %")
print(f"RMSE: {round(rmse)}")
```

```python
import pickle


model_path = "catboost_car_price_model.pkl"
with open(model_path, "wb") as file:
    pickle.dump(cat_model, file)

print(f"Model saved successfully at {model_path}")
```

```python
import pickle
from sklearn.metrics import r2_score, root_mean_squared_error
 
with open("catboost_car_price_model.pkl", "rb") as file:
    loaded_model = pickle.load(file)

print("Model loaded successfully!")


y_pred_loaded = loaded_model.predict(test_X)


r2_loaded = r2_score(test_y, y_pred_loaded)
rmse_loaded = root_mean_squared_error(test_y, y_pred_loaded)


print(f"Loaded Model R2: {round(r2_loaded * 100, 2)}%")
print(f"Loaded Model RMSE: {round(rmse_loaded)}")
```
